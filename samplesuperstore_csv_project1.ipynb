```python
REGION_SALES=df.groupby('Region')['Sales'].sum().reset_index()
REGION_SALES


plt.figure(figsize=(8,8))
sns.barplot(data=REGION_SALES,x='Region',y='Sales')
plt.title('REGION_SALES')
plt.show()
```

```python
CATEGORY_SALES=df.groupby('Category')['Sales'].sum().reset_index()

plt.figure(figsize=(8,8))
sns.barplot(data=CATEGORY_SALES,x='Category',y='Sales')
plt.title('CATEGORY_SALES')
plt.show()
```

```python
REGION_PROFIT=df.groupby('Region')['Profit'].sum().reset_index()


plt.figure(figsize=(8,8))
sns.barplot(data=REGION_PROFIT,x='Region',y='Profit')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('REGION_PROFIT')
plt.show()
```

```python
CATEGORY_PROFIT=df.groupby('Category')['Profit'].sum().reset_index()

plt.figure(figsize=(8,8))
sns.barplot(data=CATEGORY_PROFIT,x='Category',y='Profit')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.title('CATEGORY_PROFIT')
plt.show()
```

```python
sns.scatterplot(data=df,x='Discount',y='Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Discount vs Profit')
plt.show()
```

```python
SUB_CATEGORY=df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()


plt.figure(figsize=(8,8))
sns.barplot(data=SUB_CATEGORY,x='Sub-Category',y='Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.title('SUB_CATEGORY')
plt.xticks(rotation=90)
plt.show()
```
